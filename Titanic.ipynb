# 9) Handle Missing Values Handle Missing Values #

Burada missing values leri dolduracağız.

```python
data.isnull().sum() #--> Görülüdüğü üzere embarked ve age de eksik veriler var.
```

```python
data["Embarked"].mode() #--> En çok tekrar edeni bulabilmek için.
```

S Southampton u temsil etmekte.En çok tekrarlayan S imiş.O zaman 2 boş yeri S ile dolduralım.

```python
data["Embarked"].fillna("S",inplace=True) #--> Fillna fonksiyonu ile missing valuesleri doldurduk.
```

```python
data.isnull().sum()
```

Şimdi sırada yaş var .177 yeri ortalama yaş ile dolduralım.

```python
data["Age"].fillna(data["Age"].mean(),inplace=True) # inplace=True kalıcı olarak değişmesini sağlar.
```
