# 17) Group players by age (Age<18 --> Very Young) (Age<24 --->Young) (Age<32 --> MiddleAge) (Age>32 --> Old)

```python
def Age(Age):
    if Age<=20:
        return "Very Young"
    elif Age<=26:
        return "Young"
    elif Age<=33:
        return "Middle Age"
    else:
        return "Old"
```

```python
data["AgeScala"]=data["Age"].apply(Age)
```

```python
data["AgeScala"].nunique() #--->Control
```

```python
data.head(5)
```

```python
sns.barplot(x="Goals",y="AgeScala",data=data)
```
